```python
#Hello! I am an engineering student and a beginner with coding in Python. 
#For this specific project, the code will read data from a specific column,
#find the average of that column, and then output a line chart of all the
#data points as well as the average line. For my data, I have chosen to use
#two files from https://portal.aodn.org.au/search, which is ocean data from the 
#coasts of Australia. I wanted to compare data from the West and East
#coasts and find any differences between them.

import numpy as np
import matplotlib.pyplot as plt

#This code will read all data from a specified .CSV file, and since
#there is a header on each .CSV that follows the same format, there is
#a command that will skip the header and read the necessary data points
West = np.genfromtxt("CTDWest.csv", delimiter = ",", skip_header = 30)
East = np.genfromtxt("CTDEast.csv", delimiter = ",", skip_header = 30)

#These lines will find a specific column within the .CSV file and create
#an average value. In this case, column 22 is the salinity data.
Avgwsal =np.nanmean(West[:,22])
Avgesal =np.nanmean(East[:,22])

#These lines perform the same functions of the two lines above, but
#will look for column 20, which is the temperature data.
Avgwtemp =np.nanmean(West[:,20])
Avgetemp =np.nanmean(East[:,20])

#These lines will print/output the averages for both salinity and temperature
print("Average West Coast Salinity =", round(Avgwsal,2),"ppt")
print("Average East Coast Salinity =", round(Avgesal,2),"ppt")
print("Average West Coast Temperature =", round(Avgwtemp,2), "Celcius")
print("Average East Coast Temperature =", round(Avgetemp,2), "Celcius")

#The following lines of code will create 4 plots in a 2x2 pattern. They will plot
#all data points found from the chosen column as well as the average value of that data

#Salinity data for the West Coast
plt.subplot(2, 2, 1)
plt.plot(West[:,22], color = "c")
plt.axhline(Avgwsal, linestyle = "--", color = "k")
plt.title("West Coast")
plt.xlabel('Data Point #')
plt.ylabel('Salinity (ppt)')
#plt.xlim(0,800)

#Salinity data for the East Coast
plt.subplot(2, 2, 2)
plt.plot(East[:,22], color = "k")
plt.title("East Coast")
plt.axhline(Avgesal, linestyle = "--", color = "r")

#Temperature data for the West Coast
plt.subplot(2, 2, 3)
plt.plot(West[:,20])
plt.axhline(Avgwtemp, linestyle = "--", color = "k")
plt.xlabel('Data Point #')
plt.ylabel('Temperature (Celcius)')
#plt.xlim(0,800)

#Temperature data for the East Coast
plt.subplot(2, 2, 4)
plt.plot(East[:,20])
plt.axhline(Avgetemp, linestyle = "--", color = "r")
plt.xlabel('Data Point #')
```
